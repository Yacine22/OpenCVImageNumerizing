```python
import cv2 as cv
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
image = cv.imread("C:\\Users\\HP\\Pictures\\parame.jpg", 0) # Charger une image 
print(image)
print("Dimensions", image.shape) # Le format numérique de l'image ( taille très grande)
```

```python
resise = cv.resize(image,(15,15)) # réduire la taille de l'image pour bien montrer son format numérique
print(resise)
```

```python
cv.imshow("imag", resise)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
fig, ax = sns.heatmap(resise, annot=True, fmt="d")
im = ax.imshow(image)
fig.tight_layout()
plt.show()
```
